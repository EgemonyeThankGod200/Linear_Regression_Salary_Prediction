```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
Salary = pd.read_csv('Salary_dataset.csv')
Salary
```

```python
Salary_1 = Salary.drop(columns= 'Unnamed: 0')
Salary_1
```

```python
names = {'YearsExperience':'Years of Experience', 'Salary': 'Salary'}
Salary_1 = Salary_1.rename(columns= names)
Salary_1
```

```python
# feature selection
X = Salary_1.iloc[:,:-1].values
Y= Salary_1.iloc[:, 1].values
```

```python
# show values of X
X
```

```python
# show values of Y
Y
```

```python
# Training and Testing Splitting
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size =0.3, random_state=0)
```

```python
X_train, X_test, Y_train, Y_test
```

```python
# instantiating Linear regression class
Reg = LinearRegression()
# fitting split train dataset
Reg.fit(X_train, Y_train)
Y_predict = Reg.predict(X_test)
print('R2_score: ', r2_score(Y_predict, Y_test), 'Mean_Squared_Error: ', np.sqrt(mean_squared_error(Y_predict, Y_test)), 'Mean_Absolute_Error: ', mean_absolute_error(Y_predict, Y_test))
```

```python
print('Coefficent: ', Reg.coef_, 'intercept: ', Reg.intercept_)
```

```python
# Printing Y_predict to compare actual (Y_test) and predicted result (Y_predict)
Y_predict
```

```python
Y_test
```

```python
# lets try and predict salary of some years of experience not included in our data set
Salary_predict =Reg.predict([[11.0], [12.1], [20.2]])
print(f'Salary_predict is : {Salary_predict}')
```

```python
# regression plotting
sns.regplot(X_train, Y_train)
```

```python
plt.scatter(X_train, Y_train, color='red')
plt.plot(X_train, Reg.predict(X_train), color='blue')
plt.title('Salary vs Years of Experience', size=15)
plt.xlabel('Years of Experience', size =15)
plt.ylabel('Salary', size=15)
plt.show()
```

```python
plt.scatter(X_test, Y_test, color='red')
plt.plot(X_test, Reg.predict(X_test), color='blue')
plt.title('Salary vs Years of Experience', size=15)
plt.xlabel('Years of Experience', size =15)
plt.ylabel('Salary', size=15)
plt.show()
```
